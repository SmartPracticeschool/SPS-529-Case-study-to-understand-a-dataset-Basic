```python
df=pd.read_csv("C:/Users/LAKSHMI/Documents/boston_data.csv")
```

```python
df_binary=df[['crim','lstat']]
```

```python
df_binary.columns=['crime','stat']
```

```python
df_binary.head()
```

```python
X= np.array(df_binary['crime']).reshape(-1,1)
y = np.array(df_binary['stat']).reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=1/3,random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
```

```python
regressor
```

```python
regressor.fit(X_train,y_train)
```

```python
print(regressor.score(X_test,y_test))
```

```python
y_pred=regressor.predict(X_test)
x_pred=regressor.predict(X_train)
```

```python
y_pred
```

```python
x_pred
```

```python
mp.scatter(X_train,y_train,color="violet")
mp.plot(X_train,x_pred,color="black")
mp.title("Training dataset")
mp.xlabel("crim")
mp.ylabel("stat")
mp.show()
```

```python
mp.scatter(X_train,y_train,color="yellow")
mp.plot(X_train,x_pred,color="blue")
mp.title("Testing dataset")
mp.xlabel("crim")
mp.ylabel("stat")
mp.show()
```
